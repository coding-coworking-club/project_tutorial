```python
import os
import sys
from selenium import webdriver
from time import sleep
from bs4 import BeautifulSoup
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
```

```python
options = Options()
options.add_argument("--disable-notifications")
driver = webdriver.Chrome('/home/seraphine/workspace/project_tutorial/e_shopping_website/resource/chromedriver', options=options)
option = webdriver.ChromeOptions()
driver.implicitly_wait(10) # seconds
```

```python
home = "https://www.genquo.com/zh-tw/all"
driver.get(home)
print(driver.title)
print(driver.current_url)
```

```python
item_list_path = list()
for page in range(1, 44):
    #把本頁搜尋商品的網址截下來
    driver.get(f"https://www.genquo.com/zh-tw/all?P={page}")
    WebDriverWait(driver, 50).until(
    EC.presence_of_element_located((By.CLASS_NAME, "item-list"))
    )
    item_list = driver.find_element_by_class_name("item-list")
    item_list = item_list.find_elements_by_class_name("item")
    tmp_list = list()
    for item in item_list:
        item_path = item.find_element_by_class_name("item__name").find_element_by_tag_name("a").get_attribute("href").split("?")[0]
        if item_path not in tmp_list:
            tmp_list.append(item_path)
    item_list_path.extend(tmp_list)
print(item_list_path)
```

```python
#新的分頁
for path in item_list_path:
    driver.get(path)
    print(path)
    try:
        WebDriverWait(driver, 50).until(
        EC.presence_of_element_located((By.CLASS_NAME, "position-relative"))
        )
    except:
        continue
    picture_url = driver.find_element_by_class_name("position-relative").find_element_by_tag_name("img").get_attribute("src")
    print(picture_url)

    name = driver.find_element_by_class_name("product-title").text.strip()
    print(name)

    price = driver.find_element_by_class_name("product-price").text.strip().split()[0]
    print(price)
    
    color_list = driver.find_element_by_class_name("product-color").find_elements_by_tag_name("li")
    color_list = [color.find_element_by_tag_name("img").get_attribute("title") for color in color_list]
    print(color_list)

    size_select = driver.find_element_by_id("sizeSelect")
    size_select.click()
    size_list = size_select.find_element_by_tag_name("ul").find_elements_by_css_selector("li")
    size_list = [size.text for size in size_list]
    print(size_list)
    print("---------------------")
driver.quit()
```
