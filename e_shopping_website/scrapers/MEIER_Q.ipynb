```python
import time
from selenium import webdriver

driver = webdriver.Chrome('/Users/cuiyaqing/Downloads/chromedriver')  # Optional argument, if not specified will search path.
```

```python
driver.get('https://www.meierq.com/all');
```

```python
item_list = driver.find_element_by_class_name("items")
```

```python
item_number = []
for item in item_list.find_elements_by_tag_name("li"):
    number = item.find_element_by_tag_name("a").get_attribute("href").split("n")[1].split("/")[1].split("?")[0]
    if number in item_number:
        continue
    item_number.append(item.find_element_by_tag_name("a").get_attribute("href").split("n")[1].split("/")[1].split("?")[0])
```

```python
item_number = []
for page in range(1,63):
    url = f"https://www.meierq.com/all?P={page}"
    driver.get(url)
    item_list = driver.find_element_by_class_name("items")
    for item in item_list.find_elements_by_tag_name("li"):
        number = item.find_element_by_tag_name("a").get_attribute("href").split("n")[1].split("/")[1].split("?")[0]
        if number in item_number:
            continue
        item_number.append(item.find_element_by_tag_name("a").get_attribute("href").split("n")[1].split("/")[1].split("?")[0])
```
