```python
import os
import sys
from selenium import webdriver
from time import sleep
from bs4 import BeautifulSoup
from selenium.webdriver.chrome.options import Options
os.path.dirname(sys.executable)
```

```python
options = Options()
options.add_argument("--disable-notifications")
driver = webdriver.Chrome('/home/seraphine/workspace/project_tutorial/e_shopping_website/resource/chromedriver', options=options)
driver.implicitly_wait(10) # seconds
option = webdriver.ChromeOptions()
```

```python
file_path = os.path.abspath(os.getcwd())
file_path
```

```python
home = "https://www.mercci22.com/zh-tw/all"
driver.get(home)
driver.implicitly_wait(10) # seconds
print(driver.title)
print(driver.current_url)
sleep(1)
```

```python
item_list_path = list()
for page in range(1, 80):
    driver.get(f"https://www.mercci22.com/zh-tw/all?P={page}")
    driver.find_element_by_class_name("items")
    #把本頁搜尋商品的網址截下來
    soup = BeautifulSoup(driver.page_source,"html.parser")
    item_list = soup.find("ul",{"class":"items"}).find_all("a",{"class":"items__image js-loaded"})
    tmp_list = list()
    for item in item_list:
        item_path = item["href"].split("?")[0]
        if item_path not in tmp_list:
            tmp_list.append(item_path)
    item_list_path.extend(tmp_list)
```

```python
print(item_list_path)
```

```python
#新的分頁
for path in item_list_path:
    url = f"https://www.mercci22.com{path}"
    driver.get(url)
    print(url)

    picture_url = driver.find_element_by_class_name("position-relative").find_element_by_tag_name("img").get_attribute("src")
    print(picture_url)

    name = driver.find_element_by_class_name("product-title").text.strip()
    print(name)
    
    price = driver.find_element_by_class_name("product-price").text.strip()
    print(price)

    color_list = driver.find_element_by_class_name("product-color").find_elements_by_tag_name("li")
    color_list = [color.find_element_by_tag_name("img").get_attribute("title") for color in color_list]
    print(color_list)

    size_select = driver.find_element_by_id("sizeSelect")
    size_select.click()
    size_list = size_select.find_element_by_tag_name("ul").find_elements_by_css_selector("li")
    size_list = [size.text for size in size_list]
    print(size_list)
    print("---------------------")
```

```python
#新的分頁
for path in item_list_path:
    url = f"https://www.mercci22.com{path}"
    driver.get(url)
    print(url)
    
    soup = BeautifulSoup(driver.page_source,"html.parser")
    picture_url = soup.find("div",{"class":"position-relative"}).img["src"]
    print(picture_url)

    name = soup.find("h1",{"class":"product-title"}).text.strip()
    print(name)

    color_list = soup.find("div",{"class":"product-color"}).find_all("li")
    color_list = [color.img["title"] for color in color_list]
    print(color_list)

    size_list = soup.find("ul",{"class":"r-select__options"}).find_all("li")
    size_list = [size.text.strip() for size in size_list]
    print(size_list)
    print("---------------------")
```
