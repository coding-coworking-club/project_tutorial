```python
home = "http://tokichoi.91app.com/"
driver.get(home)
driver.implicitly_wait(10) # seconds
print(driver.title)
print(driver.current_url)
# sleep(1)

search_input = driver.find_element_by_class_name("nav-menu-link")
print("first text", search_input.text)
print(f"click {search_input.text}")
search_input.click()
search_input = driver.find_element_by_class_name("icon-slim-arrow-down")
# sleep(1)
print(f"click {search_input.text}")
search_input.click()
# sleep(1)
search_input = driver.find_element_by_class_name("category-level1")
search_input = driver.find_element_by_class_name("category-menu-item-link")
print("second text", search_input.text)
print(f"click {search_input.text}")
search_input.click()
# sleep(1)
```

```python
for i in range(1, 20):
    driver.execute_script("window.scrollTo(0,document.body.scrollHeight)")
    sleep(1)
```

```python
#把本頁搜尋商品的網址截下來
soup = BeautifulSoup(driver.page_source,"html.parser")
item_list = soup.find_all("li",{"class":"column-grid-container__column"})

item_list_path = [] 
for item in item_list:
    item_list_path.append(item.a["href"])

print(item_list_path)
```

```python
#新的分頁
for path in item_list_path:
    url = f"http://tokichoi.91app.com{path}"
    driver.get(url)
    print(url)
    sleep(0.5)
    #爬詳細的圖文
    soup = BeautifulSoup(driver.page_source,"html.parser")
    
#     picture = soup.find("figure",{"class":"large-image-frame"}).div.a.get("href")
    picture = soup.find("figure",{"class":"large-image-frame"}).img["src"]

    picture_url = f"https:{picture}"
    print(picture_url)
    
    picture_name = soup.find("h1",{"class":"salepage-title"}).text.strip()
    if "】" in picture_name:
        picture_name = picture_name.split("】")[1].split("(")[0]
    else:
        picture_name = picture_name.split("(")[0]
    print(picture_name)
    
    clothes_price = soup.find("div",{"class":"salepage-price cms-moneyColor"}).text
    print(clothes_price)
    
    picture_color_all = soup.find_all("ul", {"class": "sku-ul"})
    
    if picture_color_all:
        try:
            color_list, size_list = picture_color_all
            color_list = color_list.find_all("a")
            color_list = [color.text for color in color_list]
            print(color_list)
            size_list = size_list.find_all("a")
            size_list = [size.text for size in size_list]
            print(size_list)
        except Exception as e:
            print(e)
            color_list = picture_color_all[0]
            color_list = color_list.find_all("a")
            color_list = [color.text for color in color_list]
            print(color_list)
    else:
        color_list = []
        size_list = []
    print("---------------------")
```
